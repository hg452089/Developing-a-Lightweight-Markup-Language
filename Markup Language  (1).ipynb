```python
p=open ('html.txt','w')
p.write('<html>'+'\n')
dp=open ('TXT.txt','r')
x=dp.read()
val = {'$':['<b>','</b><br>'],'@':['<head>','</head><br>'],'!':['<title>','</title><br>'],'_':['<i>','</i><br>'],'#':['<h1>','</h1><br>'],'%':['<h2>','</h2><br>'],'^':['<h3>','</h3><br>'],'*':['<ul>','</ul><br>'],'-':['<ol>','</ol><br>'],'~':['<li>','</li><br>']}
#a.type()
#print(a.split('\n')) # spliting lines ending with \n.
a=x.split('.')
#print(val['$'][0])
#print(a)    
for items in a:
    try:
        def recur(st):
            if st.find('$')!=(-1):
                s1=st.replace('$',val['$'][0],1)
                s2=s1.replace('$',val['$'][1],1)
                return recur(s2)
            elif st.find('@')!=(-1):
                s1=st.replace('@',val['@'][0],1)
                s2=s1.replace('@',val['@'][1],1)
                return recur(s2)
            elif st.find('!')!=(-1):
                s1=st.replace('!',val['!'][0],1)
                s2=s1.replace('!',val['!'][1],1)
                return recur(s2)
            elif st.find('_')!=(-1):
                s1=st.replace('_',val['_'][0],1)
                s2=s1.replace('_',val['_'][1],1)
                return recur(s2)
            elif st.find('#')!=(-1):
                s1=st.replace('#',val['#'][0],1)
                s2=s1.replace('#',val['#'][1],1)
                return recur(s2)
            elif st.find('%')!=(-1):
                s1=st.replace('%',val['%'][0],1)
                s2=s1.replace('%',val['%'][1],1)
                return recur(s2)
            elif st.find('^')!=(-1):
                s1=st.replace('^',val['^'][0],1)
                s2=s1.replace('^',val['^'][1],1)
                return recur(s2)
            elif st.find('*')!=(-1):
                s1=st.replace('*',val['*'][0],1)
                s2=s1.replace('*',val['*'][1],1)
                return recur(s2)
            elif st.find('-')!=(-1):
                s1=st.replace('-',val['-'][0],1)
                s2=s1.replace('-',val['-'][1],1)
                return recur(s2)
            elif st.find('~')!=(-1):
                s1=st.replace('~',val['~'][0],1)
                s2=s1.replace('~',val['~'][0],1)
                return recur(s2)
            else:
                return p.write(st)
        recur(items)
    except:
        print("Invalid input!!!")

    else:
        dp.close()
p.write('\n'+'</html>')
p.close()
```
